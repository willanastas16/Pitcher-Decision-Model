```python
combined_df["Results"] = combined_df["BIP Result"].fillna('') + combined_df["Non-BIP Result"].fillna('')
combined_df["Results"] = combined_df["Results"].replace('', None)

combined_df.drop(columns=["BIP Result", "Non-BIP Result"], inplace=True)

combined_df.head()
```

```python
required_columns = {"Date", "Pitcher", "Hitter", "Inning"}
missing_cols = required_columns - set(combined_df.columns)
if missing_cols:
    print(f"Missing columns: {missing_cols}")
else:
    # Sort data for correct sequencing within games
    combined_df = combined_df.sort_values(by=["Date", "Pitcher", "Inning"])

    # Compute Pitch Count: Cumulative count of pitches per pitcher
    combined_df["Pitch Count"] = combined_df.groupby(["Date", "Pitcher"]).cumcount() + 1

    # Identify first pitch of each new at-bat
    combined_df["New AB"] = (combined_df["Count"] == "0-0").astype(int)  # Assuming 'Count' column tracks ball-strike count

    # Compute Pitches Per AB (Reset properly)
    combined_df["Pitches Per AB"] = combined_df.groupby(["Date", "Pitcher", "Hitter", "New AB"])["Pitch Count"].cumsum() + 1

    # Compute cumulative sum of 'New AB'
    combined_df["Cumulative New AB"] = combined_df.groupby(["Pitcher", "Date"])["New AB"].cumsum()

    # Create 'Time Through Order' based on the cumulative sum, grouped by 'Pitcher' and 'Date'
    combined_df["Time Through Order"] = (combined_df["Cumulative New AB"] // 10) + 1




    # Display results
combined_df.head()
```

```python
import numpy as np

def calculate_runs(combined_df):
    # Extract scores and split into two separate team scores
    scores = combined_df["Score"].str.split("-", expand=True).astype(int)
    
    # Compute total runs before and after the play
    runs_before = scores.sum(axis=1)
    runs_after = runs_before.shift(-1)  # Shift up to get the next row's total runs
    
    # Compute runs on the play (absolute difference)
    runs_on_play = (runs_after - runs_before).abs()
    
    # Assign calculated values back to the DataFrame
    combined_df["Runs Before"] = runs_before
    combined_df["Runs After"] = runs_after
    combined_df["Runs on the Play"] = runs_on_play.replace(0, np.nan)  # Replace 0 with NaN for clarity
    
    return combined_df

```

```python
combined_df = calculate_runs(combined_df)
combined_df
```

```python
def calculate_re24(combined_df):
    # Filter only valid outs (0, 1, 2)
    combined_df = combined_df[combined_df["Outs"].isin([0, 1, 2])]

    # Create a new column representing the Base-Out State
    combined_df["Base-Out State"] = combined_df["Runners On Base"] + "_" + combined_df["Outs"].astype(str)
    
    # Compute Run Expectancy by averaging "Runs on the Play" for each base-out state
    re24_table = combined_df.groupby("Base-Out State")["Runs on the Play"].mean().reset_index()
    
    # Rename column for clarity
    re24_table.rename(columns={"Runs on the Play": "RE24"}, inplace=True)
    
    return re24_table

# Apply function to compute RE24 values
re24_values = calculate_re24(combined_df)
print(re24_values)
```
