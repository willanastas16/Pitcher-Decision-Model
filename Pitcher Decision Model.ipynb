```python
combined_df["Results"] = combined_df["BIP Result"].fillna('') + combined_df["Non-BIP Result"].fillna('')
combined_df["Results"] = combined_df["Results"].replace('', None)

combined_df.drop(columns=["BIP Result", "Non-BIP Result"], inplace=True)

combined_df.head()
```

```python
required_columns = {"Date", "Pitcher", "Hitter", "Inning"}
missing_cols = required_columns - set(combined_df.columns)
if missing_cols:
    print(f"Missing columns: {missing_cols}")
else:
    # Sort data for correct sequencing within games
    combined_df = combined_df.sort_values(by=["Date", "Pitcher", "Inning"])

    # Compute Pitch Count: Cumulative count of pitches per pitcher
    combined_df["Pitch Count"] = combined_df.groupby(["Date", "Pitcher"]).cumcount() + 1

    # Identify first pitch of each new at-bat
    combined_df["New AB"] = (combined_df["Count"] == "0-0").astype(int)  # Assuming 'Count' column tracks ball-strike count

    # Compute Pitches Per AB (Reset properly)
    combined_df["Pitches Per AB"] = combined_df.groupby(["Date", "Pitcher", "Hitter", "New AB"])["Pitch Count"].cumsum() + 1

    # Compute cumulative sum of 'New AB'
    combined_df["Cumulative New AB"] = combined_df.groupby(["Pitcher", "Date"])["New AB"].cumsum()

    # Create 'Time Through Order' based on the cumulative sum, grouped by 'Pitcher' and 'Date'
    combined_df["Time Through Order"] = (combined_df["Cumulative New AB"] // 10) + 1




    # Display results
combined_df.head()
```
